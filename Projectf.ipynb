```python
# The libraries 

import pandas as pd 
import numpy as np 
import plotly.express as px 
import plotly.graph_objects as go # Graphs other than scatter and bar or customize
```

```python
# Save the file path 
p1 = '/content/IED.xlsx'

#Read file 

ied= pd.read_excel(p1)

#Check 

ied.info()
```

```python
ied = ied.sort_values(by='Year')
ied.head()
```

```python
# Drop does not delete date
ied = ied.set_index('Year', drop = False)
ied.head()
```

```python
px.line(ied['Sp'],
        animation_frame = ied['Year'].astype(str),
        range_x =['1992','2021'],
        range_y =[-1000,12000])
```

```python
fig = go.Figure(
    layout = go.Layout(
        updatemenus = [dict(type= "buttons", direction = "right", x = 1, y = 1.25)],
        xaxis = dict(range = ['1993','2022'],
                     autorange = False),
        yaxis = dict(range = [-1000,12000],
                     autorange = False, 
                     title_text = ''),
                     title = 'Foreign direct investment in Colombia by Sector',
        title_font_size = 25,
        title_x = 0.5
        )
)
```

```python
fig.show()
```

```python
#Animation

fig.update(frames = frames)
```

```python
# ADD MORE FORMATING 

fig.update_xaxes(title='' , title_font_size = 20,
                 dtick = 'M3',
                 tickformat = '%Y')
fig.update_yaxes(nticks=15)
fig.update_layout(legend=dict(
                  orientation = "h",
                  yanchor = "middle",
                  y=-0.2,
                  xanchor = "center",
                  x=0.47))

fig.update_layout(font=dict(family="Courier New"))

fig.update_layout(template = 'plotly_dark')

```

```python
# Buttons 

fig.update_layout(
    updatemenus = [
        dict(
            buttons = list([
                dict(label = 'Play',
                     method = 'animate',
                     args = [None,{'frame':{'duration':500}}]
                     )
                 ]
            )
        )
    ]
)

# The result

fig.show()
```
